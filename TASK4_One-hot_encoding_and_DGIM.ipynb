```python
# SMALL EXAMPLE OF THE Dgim PACKAGE
from dgim import Dgim
dgim = Dgim(N=32, error_rate=0.1)
for i in range(100):
    dgim.update(True)
dgim_result = dgim.get_count() # 30 (exact result is 32)
dgim_result
```

```python
# APPLIED ON OUR DATA SET
dgim = Dgim(N=1385, error_rate=0.1)

for i in range(len(chosen_column)):
    if chosen_column[i] == 1:
        dgim.update(True)
    else:
        dgim.update(False)
    
dgim_result = dgim.get_count()
dgim_result
```
